```python
# ==========================================
# 独立画图脚本：基于 Excel 数据重绘 Loss 和 PPL 曲线
# ==========================================
import pandas as pd
import matplotlib.pyplot as plt
import os

# 1. 设定云盘路径和 Excel 文件路径
CLOUD_DISK_PATH = "./poetry_output"
excel_path = os.path.join(CLOUD_DISK_PATH, "training_data_report.xlsx")

print(f"📂 正在读取 Excel 数据: {excel_path} ...")

# 2. 读取 Excel 数据
# 如果提示缺少 openpyxl，可以运行 !pip install openpyxl
df = pd.read_excel(excel_path)

# 提取列数据
epochs_list = df['Epoch']
train_loss = df['Train_Loss']
val_loss = df['Val_Loss']
train_ppl = df['Train_PPL']
val_ppl = df['Val_PPL']

# 设置画图风格
plt.style.use('seaborn-v0_8-whitegrid')

# ==========================================
# 📊 绘制并保存 Loss 曲线
# ==========================================
plt.figure(figsize=(8, 5), dpi=150)
plt.plot(epochs_list, train_loss, marker='o', label='Train Loss')
plt.plot(epochs_list, val_loss, marker='s', label='Validation Loss')
plt.title('Model Training and Validation Loss', fontsize=14)
plt.xlabel('Epochs', fontsize=12)
plt.ylabel('Cross Entropy Loss', fontsize=12)
plt.legend()
plt.tight_layout()

# 保存并显示
loss_img_path = os.path.join(CLOUD_DISK_PATH, 'report_loss_curve_redrawn.png')
plt.savefig(loss_img_path, dpi=300)
plt.show()  # 直接在网页里展示出来
plt.close()

# ==========================================
# 📈 绘制并保存 PPL 曲线 (带对数坐标)
# ==========================================
plt.figure(figsize=(8, 5), dpi=150)
plt.plot(epochs_list, train_ppl, marker='^', color='green', label='Train PPL')
plt.plot(epochs_list, val_ppl, marker='v', color='red', label='Validation PPL')
plt.title('Model Perplexity (PPL)', fontsize=14)
plt.xlabel('Epochs', fontsize=12)
plt.ylabel('Perplexity (Log Scale)', fontsize=12)
plt.legend()
plt.tight_layout()

# 灵魂一步：设置纵坐标为对数刻度
plt.yscale('log')

# 保存并显示
ppl_img_path = os.path.join(CLOUD_DISK_PATH, 'report_ppl_curve_redrawn.png')
plt.savefig(ppl_img_path, dpi=300)
plt.show()  # 直接在网页里展示出来
plt.close()

print("✅ 重绘完成！新的图片已在上方显示，并保存为带有 '_redrawn' 后缀的文件。")
```
